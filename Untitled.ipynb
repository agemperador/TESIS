```python
def get_2d_y_var(y,p, x_corte):
    
    y = y[:,x_corte]
    
    press = np.array(p[:,x_corte,x_corte]) 
    
    y2D, press2D = np.meshgrid( y, press)

    return y2D, press2D
    
def get_lat_somb (data,var_s, x_corte, press2D, mask_s):
        
    y = data.variables['XLAT'][10,:,x_corte]
    topo = data.variables['HGT']
    topo = np.asarray(list(map(lambda x: -1.225*9.8*x +101300 ,topo)))
    
    var_s = np.array(var_s)[:-1,:, x_corte]   

    for j in range(305):
        
        top = topo[10,j,x_corte]
                                
        var_s[:,j] = np.ma.array(var_s[:,j], mask = press2D[:,j] > top)
        
    var_s = np.ma.array(var_s , mask = np.abs(var_s) < mask_s)
    
    return var_s
```

```python
def collage(data, carpeta,bm,lon,lat, x,y, max_lat,max_lon, hora, hora_final,topo, r_horaria, w,u,v,t ,q, cmap_lluvia, guardar = False, show = True):
   


    hora = 4*hora
    hora_final = hora_final*4

    window = 20
    wi = np.zeros((11,len(w[0,:,1,1])))

    max_lats = np.zeros(11)
    max_lons = np.zeros(11)

    stride = 11
    stride_lon = 16

    vmin_lluvia = 5
    vmax_lluvia = 50
    vmin_w = -20
    vmax_w = 20

    cmap_lluvia = 'gist_ncar'

    wcmap = 'coolwarm'

    w_level = 18

    ix_sc = []

    for i in range(hora,hora+hora_final):

        """"""""""""""""""
######## Deteccin de mximos    
########
        """"""""""""""""""

        rsample = r_horaria[i]
        wsample = w[i,...]
        rain = np.ma.array(rsample,mask = rsample <3 ) 


        if max_lat or max_lon:
               
                wminisample = wsample[:,max_lat-stride:max_lat+stride,max_lon-stride_lon:max_lon+stride_lon]
               
                maximo = np.argwhere(wminisample[w_level,:,:] == wminisample[w_level,:,:].max())[0]

                max_lat_pr = max_lat - stride
                max_lon_pr = max_lon - stride_lon 

        else: 
            maximo = np.argwhere((wsample[w_level,:,:] == wsample[w_level,:,:].max()))[0]
            max_lat_pr = 0
            max_lon_pr = 0


        max_lat =  maximo[0] + max_lat_pr
        max_lon =  maximo[1] + max_lon_pr

        if max_lon + window >= x.shape[0]: 
            max_lat = 0
            max_lon = 0
            continue

        ix_sc.append((max_lat,max_lon))

        xi = x[max_lat-window:max_lat+window,max_lon-window:max_lon+window]
        yi = y[max_lat-window:max_lat+window,max_lon-window:max_lon+window]



#######################################################
###########            ACA EMPIEZA LA FIGURA         ##
#######################################################

        fig = plt.figure(figsize=(14,15))
        gs = fig.add_gridspec(5, 2)


        """"""
####### SLO LA TORMENTA
#######
        """"""
################
##------> EJE 1
################

        ax0 = fig.add_subplot(gs[0:1,0])

        plt.title('Zoom de la tormenta trackeada viento en 500hPa')


        pq =  ax0.pcolormesh(xi,yi,q[i,0,max_lat-window:max_lat+window,max_lon-window:max_lon+window], cmap = plt.get_cmap('BrBG'),zorder = -2)

        ps = ax0.pcolormesh(xi,yi,rain[max_lat-window:max_lat+window,max_lon-window:max_lon+window],
                            cmap=plt.get_cmap(cmap_lluvia), vmin =  vmin_lluvia, vmax=vmax_lluvia, zorder = 2)  

        ax0.contour(xi,yi,topo[0,max_lat-window:max_lat+window,max_lon-window:max_lon+window], 
                    colors = 'black', linewidth = 0.5, zorder = 1,alpha = 0.5)

        ax0.scatter(xi[window,window],yi[window,window],c= 'orange',zorder = 3)

        var_c = w[i,w_level,max_lat-window:max_lat+window,max_lon-window:max_lon+window]

        ncont = 5
        c_min, c_max =np.round(np.min(var_c)),np.round(np.max(var_c))
        c_m = np.max((abs(c_min),abs(c_max)))

        if c_min != c_max:
            clev=np.linspace(-c_m,c_m,ncont)
            clev= np.delete(clev,np.argwhere(clev == 0))
            pc = ax0.contour(xi,yi,var_c,clev, cmap = plt.get_cmap('seismic') )

            clab = ax0.clabel(pc,clev,fontsize=12,fmt='%.0f',colors='k')


        var_u0 = u[i,17,2+max_lat-window:max_lat+window-2,2+max_lon-window:max_lon+window-2]
        var_v0 = to_np(v)[i,17,2+max_lat-window:max_lat+window-2,2+max_lon-window:max_lon+window-2]

        speed = np.sqrt((var_u0*2) ** 2 + (var_v0*2) ** 2)
        lw = 2*speed / speed.max()

        v_plot0 = plt.streamplot(xi[window,2:-2],yi[2:-2,window],var_u0,var_v0
                                 ,arrowstyle ='->',linewidth=lw, color ='k', density=0.8)


        ax0.hlines(yi[window-1,window-1],xi[0,0],xi[-1,-1], linestyles = 'dashed'
                   ,colors='r',label='Corte vertical')

        plt.scatter(xi[window-1,window-1],yi[window-1,window-1])


        plt.xticks([])
        plt.yticks([])


        """"""
####### COMPLETO CON LLUVIA Y TOPOGRAFA
#######
        """"""

#################
###------> EJE 2
#################
        ax1 = fig.add_subplot(gs[0:1,1])

        ax1.contour(x,y,topo[i], colors = 'black', linewidth = 0.5, zorder = -2)

        plt.tick_params(
            axis='y',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            left=False,
            labelleft=False,
        )

        ax1.pcolormesh(x,y,rain, cmap=plt.get_cmap(cmap_lluvia), vmin = vmin_lluvia, vmax = vmax_lluvia)
        col = 'orange'
        for la,lo in ix_sc[::-1]:

            ax1.scatter(x[la,lo], y[la,lo],c = col)

            col = 'r'

        lower_left = (lon[max_lat-window,max_lon-window],lat[max_lat-window,max_lon-window])
        lower_right =(lon[max_lat-window,max_lon+window],lat[max_lat-window,max_lon+window])
        upper_left = (lon[max_lat+window,max_lon-window],lat[max_lat+window,max_lon-window])
        upper_right =(lon[max_lat+window,max_lon+window],lat[max_lat+window,max_lon+window]) 


        xs = [lower_left[0], upper_left[0],
              upper_left[0], upper_right[0],
              upper_right[0], lower_right[0],
              lower_right[0], lower_left[0]]
        ys = [lower_left[1], upper_left[1],
              upper_left[1], upper_right[1],
              upper_right[1], lower_right[1],
              lower_right[1], lower_left[1]]

        bm.plot(xs, ys, latlon = True)    


        var_u1 = u[i,0,:,1:]
        var_v1 = v[i,0,1:,:]

        speed = np.sqrt((var_u1*2) ** 2 + (var_v1*2) ** 2)
        lw = 2*speed / speed.max()

        v_plot1 = plt.streamplot(x[max_lat,:],y[:,max_lon],var_u1,var_v1
                                 ,arrowstyle ='->',linewidth=lw, color ='g', density=0.8)




        bm.drawstates(linewidth=1)
        bm.drawparallels(np.linspace(-35.,-30,6),labels=[1,0,0,0],color='gray')
        bm.drawmeridians(np.linspace(-66,-61,6),labels=[0,0,0,1], color = 'gray')

        plt.title('Trackeo de la celda - Coords: %f Lat , %f Lon'%(lat[max_lat,max_lon],lon[max_lat,max_lon]))


        """"""
####### CORTE VERTICAL COMPLETO
#######
        """"""
################
##------> EJE 3
################
        ax2 = fig.add_subplot(gs[1:3,:])


        end = 59
        x2D,press2D,var_titae,var_v = set_corte_vert(data,max_lat,i,topo = topo, var_c = v,var_s = t[:,:end,...],press = press, mask_s=0)
        x2D = x2D[:end,max_lon-window:max_lon+window]

        press2D = press2D[:end,max_lon-window:max_lon+window]

        var_titae = var_titae[:end,max_lon-window:max_lon+window]
        var_titae = np.ma.masked_array(var_titae, press2D>=topo[i,max_lat,max_lon-window:max_lon+window])
        var_titae = np.ma.masked_array(var_titae, press2D<= 10000)
        var_titae = var_titae + 290

        var_v = var_v[:end,max_lon-window:max_lon+window]

        var_w = np.ma.masked_array(w[i,:end,max_lat,max_lon-window:max_lon+window], press2D>topo[i,max_lat,max_lon-window:max_lon+window])

        var_ws = np.ma.masked_array(var_w, abs(var_w)<2)

        red = (70/255, 244/255)
        green = (130/255,164/255)
        blue =(180/255,96/255)

        cdict = {
        'blue': [(0,0,1),(0.4,blue[0],blue[0]), (1, 0.95, 1)],
        'green': [(0,0,1),(0.4,green[0],green[0]), (1, 0.99, 0)],
        'red': [(0,0,1),(0.4,red[0],red[0]) , (1,0.99, 1)]}
        cmap_per = mpl.colors.LinearSegmentedColormap('my_colormap',cdict,25)


        q_plot = ax2.pcolormesh (x2D,press2D,var_titae, cmap = cmap_per,vmin = 298, vmax = 312)

        ascensos = ax2.pcolormesh(x2D,press2D,var_ws,cmap=plt.get_cmap('seismic'),vmin = -25,vmax=25)


        ncont = 9
        c_min, c_max =np.round(np.min(var_v)),np.round(np.max(var_v))
        if c_min != c_max:
            clev=np.linspace(c_min,c_max,ncont)
            clev= np.delete(clev,np.where(clev==0))
            pc = ax2.contour(x2D,press2D,var_v,clev, colors = 'k', zorder = 3)

            clab = ax2.clabel(pc,clev,fontsize=12,fmt='%.0f',colors='k')


        var_u = u[i,:end,max_lat,max_lon-window:max_lon+window]
        var_u = np.ma.masked_array(var_u, press2D>topo[i,max_lat,max_lon-window:max_lon+window])

        top = topo[i,max_lat,max_lon-window:max_lon+window]
        w_p = var_w
        u_p = var_u


        a=4

        speed = np.sqrt((u_p[:end,::a]*2) ** 2 + (w_p[:end,::a]*2) ** 2)
        lw = speed / speed.max()

        end_v = 30


        M = np.hypot(u_p[:end_v,::a], w_p[:end_v,::a])

        plot_v = ax2.barbs(x2D[:end_v,3:-3:a],press2D[:end_v,3:-3:a], u_p[:end_v,3:-3:a],w_p[:end_v,3:-3:a],
                           color ='#393939' ,
                           pivot = 'middle', 
                           length = 7, linewidth =1 ) 



        ax2.plot(x2D[0,:],topo[0,max_lat,max_lon-window:max_lon+window],linewidth = 5, color = 'k' )
        plt.ylim (96600.,10000)
        plt.ylabel('Presin (hPa)')

        plt.title('Corte vertical: W, Qv (somb) - Viento uw y V cont - Latitud: %s - Longitud: %s'%(str(max_lat),str(max_lon)))

        """"""
####### HODGRAFA
#######
        """"""
###############
###---> EJE 4
###############
    
    
        ax4 = fig.add_axes([0.105,0.56,0.14,0.12])

        ini = 8

        end = 28

        ix_lat = max_lat
        ix_lon = max_lon



        v_c = v[i,ini:end,ix_lat,ix_lon]
        u_c = u[i,ini:end,ix_lat,ix_lon]

        cm = plt.cm.get_cmap('BuPu')
        colors=[cm(1.*k/(end-ini)) for k in range(end-ini)]

        ax4.spines['left'].set_position('center')
        ax4.spines['bottom'].set_position('center')


        # Eliminate upper and right axes
        ax4.spines['right'].set_color('k')
        ax4.spines['top'].set_color('k')
        ax4.spines['left'].set_color('k')
        ax4.spines['bottom'].set_color('k')


        # Show ticks in the left and lower axes only
        ax4.xaxis.set_ticks_position('bottom')
        ax4.yaxis.set_ticks_position('left')

        vmax = np.max(abs(v_c))+2
        umax = np.max(abs(u_c))+2

        plt.xlim(-umax, umax)
        plt.ylim(-vmax,vmax)

        ax4.plot(u_c,v_c,color = 'k', linewidth = 2)

        xy = range(ini,end)

        sc = ax4.scatter(u_c,v_c, c =xy, cmap = cm, zorder = 5 )

        ax4.set_facecolor('gray')
        ax4.grid()

        clevs_hod = np.round(press[i,ini:end, max_lat,max_lon])

###############
##-----> EJE 5
###############
        
        ax5 = fig.add_subplot(gs[3:,:])
        
        y2D, press2D = get_2d_y_var(lat,press[i,...], x_corte = max_lon)
        
        y2D = y2D[:,max_lat-window:max_lat+window]
        press2D = press2D[:,max_lat-window:max_lat+window]
        
        var_s1 = get_lat_somb(data, w[i,...], max_lon, press2D, mask_s=0.9)
        
        var_s1 = var_s1[:,max_lat-window:max_lat+window]
        
        topo_z_y = topo[0,max_lat-window:max_lat+window,max_lon]
                        
        ax5.pcolormesh(y2D,press2D,var_s1,cmap= plt.get_cmap('seismic'))

        plt.plot(y2D[0,...],topo_z_y,linewidth = 5, color = 'k' )
        

        plt.ylim (96600.,10000)

        
        """"""
#######  EJES DE LAS COLORBAR
#######
        """"""

        caxr = fig.add_axes([0.83, 0.66, 0.02, 0.23])

        cbar_r = plt.colorbar(ps,ax = [ax0,ax1] ,cax = caxr,orientation = 'vertical',pad = 0.03)
        cbar_r.ax.set_ylabel('Precipitacin hr')

        caxw = fig.add_axes([0.13, 0.06, 0.65, 0.02])

        cbar_w = plt.colorbar(ascensos,ax=ax2,cax = caxw,orientation = 'horizontal')
        cbar_w.ax.set_xlabel('W m/s')

        caxq = fig.add_axes([0.83, 0.11, 0.02, 0.45])


        cbar_q = plt.colorbar(q_plot,ax=ax2,cax= caxq ,orientation = 'vertical')
        cbar_q.ax.set_ylabel('qvapor kg/kg')


        caxhod = fig.add_axes([0.0,0.55,0.015,0.14])

        ticks_hod = np.linspace(0,end,15)
        
        cbar_hod = plt.colorbar(sc,ax = ax4, cax = caxhod,orientation = 'vertical')
        cbar_hod.ax.set_yticklabels([np.round(press[i,ini,max_lat,max_lon]/100),np.round(press[i,ini+(end-ini)//4,max_lat,max_lon]/100),
                                     np.round(press[i,ini+2*(end-ini)//4,max_lat,max_lon]/100),np.round(press[i,end,max_lat,max_lon]/100)])
        cbar_hod.ax.set_ylabel('P (hPa)')


        """"""
####### INFO DEL PLOT Y SAVEFIG
#######
        """"""


        tiempo = ''.join(str(tiempos[i]).replace('b','').replace("'",'').replace('[','').replace(']','').replace('\n','').replace(' ',''))

        plt.suptitle('Trackeo de Celdas - %s - %s - SC 3'%(carpeta,tiempo), fontsize = 24)

        fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.8,
                        wspace=0.02, hspace=0.2)


        if guardar == True:
            plt.savefig('./img/%s/Track/SC3/Track_%s_%s_SC3.png'%(carpeta,carpeta,tiempo))

        if show == True:
            plt.show()
            
    plt.close()
    data.close()
        
```
